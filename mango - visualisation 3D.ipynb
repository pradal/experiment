Mango visualisation in 3D
==========================

The **Mango** tree is stored in the MTG format. Some are measured, others are simulated. 

The aim of this tutorial is to visualize it as a **2D graph** or as a **3D mockup**.

```python
%gui qt
%pylab inline
from vpltkdisplay import *
```

```python
from mangotree.mango3d import get_mtg, choose_tree, plot3d_with_time
from openalea.mtg import color
g = get_mtg()
```

```python
# MTG contains 4 varieties and 5 trees for each variety
varieties = ['cogshall', 'irwin', 'kegsingtonpride', 'jose']

# To select one plant in the mtg: 0 <= index < 5
plant_id = choose_tree(g, variety='cogshall', index=0)
```

```python
# define a visitor
max_order=9
def mango_visitor(g, v, turtle, time, *args, **kwds):
    insertion_angle = 45., 60.
    iangle = lambda x: insertion_angle[0] + (insertion_angle[1]-insertion_angle[0])*x
    
    
    order = g.property('order')

    if g.edge_type(v) == '+':
        p = g.parent(v)
        chs = list(g.children(p))
        index = chs.index(v)
        angle = index*360./len(chs)
        turtle.rollL(angle)
        turtle.down(iangle(order[v]/float(max_order)))

    turtle.setId(v)
    node = g.node(v)
    length = node._length
    diam = node._diameter
    y = node.year
    #if y:  turtle.setColor(node.year)

    #turtle.setColor(color_index.get(g.property(property_name).get(v,0),0))
    turtle.setWidth(diam/2.)
    turtle.rollL()
    turtle.F(length)
    
```

```python
# Apply a colormap based on a property
g = color.colormap(g, 'date_order')
```

```python
scene = plot3d_with_time(g, plant_id, time=12, visitor=mango_visitor)
PlantGL(scene)
```

```python
g.property_names()
g.plot_property('_length')

```

```python
for t in range(0,30,3):
    scene = plot3d_with_time(g, plant_id, time=t, visitor=mango_visitor)
    display(PlantGL(scene))
```
